### Importing Library

```python
import pandas as pd
import numpy as np 
import seaborn as sns 
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
label  = LabelEncoder()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
logestic = LogisticRegression()
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
from sklearn.metrics import accuracy_score ,precision_score,recall_score
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
import warnings
warnings.filterwarnings("ignore")
```

### Loading Dataset

```python
churn = pd.read_csv("churn.csv")
```

```python
churn.dtypes
## Inference -- 
      #1. We have to change Data type of SeniorCitizen into object or categorical data 
      #2. We have to change Data type of Total Charges into numeric data type 
```

```python
# Changing Datatype -- 
churn["gender"] = churn["gender"].astype("object")
churn["TotalCharges"] = pd.to_numeric(churn["TotalCharges"],errors="coerce")
   ## Inferene -- 
        #1. Data types of gender and Total charges has been changed 
        #2 . But due  to changes in data type it can happen that null values will enter in our dataset .
```

```python
churn.isnull().sum()
 ## Inference - 
    # Null values are present in Total Charges
```

```python
# Checking Distribution of Total charges
print(churn["TotalCharges"].skew())
sns.distplot(churn["TotalCharges"])
plt.show()
## Inference--
    # It is more like normal distributed so will remove null values by replacing it with mean value of Total Charges. 
```

```python
# Removing Null Value from  Total Charges--
churn["TotalCharges"] = churn["TotalCharges"].fillna(churn["TotalCharges"].mean())
```

```python
# Null values are removed. 
churn.isna().any()
```

```python
# Dividing data in two different data frame to get better result. 
cat_col = []
num_col = []
for i in churn:
    if(churn[i].nunique()<6):
        cat_col.append(churn[i])
    else:
        num_col.append(churn[i])
cat_col = pd.DataFrame(cat_col).T
num_col = pd.DataFrame(num_col).T
```

```python
cat_col.head()
```

```python
num_col.head()
```

## Univariate Analysis

```python
j = 1 
plt.figure(figsize=(10,40))
for i in cat_col:
    plt.subplot(9,2,j)
    sns.countplot(data=cat_col,x=i)
    j = j+1
plt.show()
```

## Bivariate analysis

```python

j=1 
plt.figure(figsize=(15,50))
for i in cat_col:
    plt.subplot(9,2,j)
    sns.histplot(data=cat_col,x=i,hue="Churn")
    plt.title(i)
    j=j+1
plt.show()
```

```python
 num_col.head()
    # Will remove customerID as it is of no use because it is different for each rows and making no effect on churn which is our
    # dependent variable.
```

```python
# Will do label encodig on categorical Data set . 
for i in cat_col:
    cat_col[i]= label.fit_transform(cat_col[i])
```

```python
# Label encoding implemented successfully
cat_col.head()
```

```python
# will make new data and combined with  cat_col  and num_col--
new_churn = pd.concat([cat_col,num_col],axis=1)
```

```python
new_churn.head()
```

```python
# checking outliers - 
for i in num_col:
    sns.boxplot(data=num_col,x=i)
    plt.show()
## Inference-- 
    # No outliers are present in numerical Dataset
```

### Modeling

```python
# Spliting Dataset into x and y 
y = new_churn["Churn"]
x = new_churn.drop("Churn",axis=1)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

###  Logistic Regression

```python
logestic.fit(x_train,y_train)
logestic.predict(x_test)
```

```python
# checking accuracy of logestic  the model  
print("accuracy of logestic model is",accuracy_score(y_test,logestic.predict(x_test)))
print("precision score ", precision_score(y_test,logestic.predict(x_test)))
print("recall score  ",recall_score(y_test,logestic.predict(x_test)))
```

###  Decision Tree

```python
# Fitting Decision Tree model 
dtc.fit(x_train,y_train)
dtc.predict(x_test)
```

```python
# checking accuracy of decsion Tree model  the model  
print("accuracy of Decision Tree is",accuracy_score(y_test,dtc.predict(x_test)))
print("precision score ", precision_score(y_test,dtc.predict(x_test)))
print("recall score  ",recall_score(y_test,dtc.predict(x_test)))
```

###  Random Decision Forest

```python
# Fitting  Random Decision Forest
rfc.fit(x_train,y_train)
rfc.predict(x_test)
```

```python
# checking accuracy of  Random Decision Forest  the model  
print("accuracy of  Random Decision Forest",accuracy_score(y_test,rfc.predict(x_test)))
print("precision score ", precision_score(y_test,rfc.predict(x_test)))
print("recall score  ",recall_score(y_test,rfc.predict(x_test)))
```
